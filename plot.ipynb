```python
from sympy import symbols, Eq, solve

def find_root(gamma, fake):
    a = 1+ gamma*fake
    x = symbols('x')
    equation = x**(gamma + 1) - a * x + a - 1
    

    def has_root(f, low, high):
        return f.subs(x, low) * f.subs(x, high) < 0

    def bisection(f, low, high, tol=1e-7):
        if not has_root(f, low, high):
            return None

        while high - low > tol:
            mid = (low + high) / 2.0
            # print(low, high)
            if f.subs(x, mid) == 0:
                return mid
            elif f.subs(x, low) * f.subs(x, mid) < 0:
                high = mid
            else:
                low = mid
        return mid


    root = bisection(equation, 0, 0.99)
    print(root)
    return root

find_root(32,  0.39384780582635903)
```
