```python
import matplotlib.pyplot as plt
import pandas as pd

# Reading the data from the CSV files
df_cache = pd.read_csv('report/benchmark_cache.csv')
df_flash = pd.read_csv('report/benchmark_flash.csv')

# Preparing data for plotting
# Without Flash-Attn (varying 'input')
x_list_input = df_cache['prefill'].unique().tolist()
y_lists_cache = [df_cache[df_cache['input'] == input_value]['latency'].tolist() for input_value in [1, 2, 4, 8, 16, 32, 64, 128, 256]]

# With Flash-Attn (varying 'input')
y_lists_flash = [df_flash[df_flash['input'] == input_value]['latency'].tolist() for input_value in [1, 2, 4, 8, 16, 32, 64, 128, 256]]

# Without Flash-Attn (varying 'prefill')
x_list_prefill = [1, 2, 4, 8, 16, 32, 64, 128, 256, 512, 1024]
y_lists_prefill_cache = [df_cache[df_cache['prefill'] == prefill]['latency'].tolist() for prefill in [1, 1000, 2000, 4000, 8000, 16000, 32000, 50000]]

# With Flash-Attn (varying 'prefill')
y_lists_prefill_flash = [df_flash[df_flash['prefill'] == prefill]['latency'].tolist() for prefill in [1, 1000, 2000, 4000, 8000, 16000, 32000, 50000]]

# Creating subplots
fig, axs = plt.subplots(2, 2, figsize=(15, 10))

# Plotting - Without Flash-Attn (varying 'input')
for i, y_list in enumerate(y_lists_cache):
    axs[0, 0].plot(x_list_input, y_list, label=f'input={2**i}', marker='o')
axs[0, 0].set_title('Without Flash-Attn (varying input)')
axs[0, 0].legend()
axs[0, 0].grid()

# Plotting - With Flash-Attn (varying 'input')
for i, y_list in enumerate(y_lists_flash):
    axs[0, 1].plot(x_list_input, y_list, label=f'input={2**i}', marker='o')
axs[0, 1].set_title('With Flash-Attn (varying input)')
axs[0, 1].legend()
axs[0, 1].grid()

# Plotting - Without Flash-Attn (varying 'prefill')
for i, y_list in enumerate(y_lists_prefill_cache):
    axs[1, 0].plot(x_list_prefill[:len(y_list)], y_list, label=f'prefill={[1, 1000, 2000, 4000, 8000, 16000, 32000, 50000][i]}', marker='o')
axs[1, 0].set_title('Without Flash-Attn (varying prefill)')
axs[1, 0].legend()
axs[1, 0].grid()

# Plotting - With Flash-Attn (varying 'prefill')
for i, y_list in enumerate(y_lists_prefill_flash):
    axs[1, 1].plot(x_list_prefill[:len(y_list)], y_list, label=f'prefill={[1, 1000, 2000, 4000, 8000, 16000, 32000, 50000][i]}', marker='o')
axs[1, 1].set_title('With Flash-Attn (varying prefill)')
axs[1, 1].legend()
axs[1, 1].grid()

# Adjust layout
plt.tight_layout()

plt.show()
```
